```python
# importing pandas libraries and
# glob module
import pandas as pd
```

```python
import os

# folder path
dir_path = r'/content/drive/MyDrive/all_pool_uniswap/Arb/2021_2022'

# list file and directories
res = os.listdir(dir_path)
print(res)
```

```python
# Function to recursively find Excel files in a folder and its subfolders
def find_excel_files(root_dir):
    excel_files = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.xlsx'):
                excel_files.append(os.path.join(root, file))
    return excel_files
# Find all Excel files in the specified folder and its subfolders
excel_files = find_excel_files(dir_path)
print(excel_files)
```

```python
df1 = pd.read_excel("/content/drive/MyDrive/all_pool_uniswap/Arb/2021_2022/2022/ARB-130.xlsx")
df1.head()
```

```python
# Initialize an empty DataFrame to merge the Excel data
df = pd.DataFrame()

from tqdm import tqdm
# Iterate through the Excel files and merge them into the DataFrame
for excel_file in tqdm(excel_files):
    df_temp = pd.read_excel(excel_file)
    df = pd.concat([df, df_temp], ignore_index=True, sort=False)

# Now, 'df' contains the merged data from all Excel files
```

```python
df_Arb = df
df_Arb.tail()
```

```python
df_Arb.to_csv("Arb.csv")
```
