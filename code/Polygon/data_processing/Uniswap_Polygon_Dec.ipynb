```python
import os

# folder path
dir_path = r'/content/drive/MyDrive/all_pool_uniswap/Poly/December'

# list file and directories
res = os.listdir(dir_path)
print(res)
```

```python
df1 = pd.read_excel(dir_path+"/"+"uniswap_polygon_2022_12_01.xlsx")
df1.head()
```

```python
df1.tail()
```

```python
files = res
```

```python
df =df1
from tqdm import tqdm
for file in tqdm(files): 
    # combining multiple excel worksheets 
    # into single data frames
    df = pd.concat([df, pd.read_excel(dir_path+"/"+file)],
                   ignore_index=True, sort=False)
```

```python
df_Poly_Dec = df
df_Poly_Dec.tail()
```

```python
df_Poly_Dec.to_csv("Poly_Dec.csv")
```
