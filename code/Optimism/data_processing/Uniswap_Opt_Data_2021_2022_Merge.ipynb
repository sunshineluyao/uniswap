```python
# Function to recursively find Excel files in a folder and its subfolders
def find_excel_files(root_dir):
    excel_files = []
    for root, dirs, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.xlsx'):
                excel_files.append(os.path.join(root, file))
    return excel_files
# Find all Excel files in the specified folder and its subfolders
excel_files = find_excel_files(dir_path)
print(excel_files)
```

```python
import pandas as pd
import os

# Check if "Opt.csv" file exists
if os.path.isfile("Opt.csv"):
    # If the file exists, load it into a DataFrame
    df = pd.read_csv("Opt.csv")

    # Define df_Opt by selecting specific columns
    df_Opt = df[["Timestamp", "Amount USD (Excel)"]]
else:
    # If the file doesn't exist, initialize an empty DataFrame
    df_Opt = pd.DataFrame()

# Now, you can use df_Opt in your script, whether it's loaded from the file or empty.
df_Opt
```

```python
# Continue or Update
import pandas as pd
from tqdm import tqdm
import pickle

# Load the previously saved merged_files list
with open("merged_files.pkl", "rb") as file:
    merged_files = pickle.load(file)
print(merged_files)


# Get the list of remaining files by excluding already merged ones
remaining_files = [file for file in excel_files if file not in merged_files]
print(remaining_files)
# Create a progress bar using tqdm
with tqdm(total=len(remaining_files), desc="Merging Excel Files") as pbar:
    for excel_file in remaining_files:
        df_temp = pd.read_excel(excel_file, engine='openpyxl')
        df_temp = df_temp[["Timestamp", "Amount USD (Excel)"]]
        df_Opt = pd.concat([df_Opt, df_temp], ignore_index=True, sort=False)

        # Update the lists of merged and remaining files
        merged_files.append(excel_file)
        with open("merged_files.pkl", "wb") as file:
          pickle.dump(merged_files, file)
        df_Opt.to_csv("Opt.csv", index=False)
        # Update the progress bar
        pbar.update(1)

# Save the final merged DataFrame to a CSV file
print("Merging completed. Progress saved.")
```
