```python
# importing pandas libraries and
# glob module
import pandas as pd
```

```python
#@title path
data = "/content/drive/MyDrive/DEX_Uniswap_Layer2/Eth/Amount" #@param {type:"string"}
```

```python
import os

# folder path
dir_path = r'/content/drive/MyDrive/DEX_Uniswap_Layer2/Eth/Amount'

# list file and directories
res = os.listdir(dir_path)
print(res)
```

```python
df1 = pd.read_csv(dir_path+"/"+"Eth_Amount_Jan.csv",index_col="Unnamed: 0")
df1.head()
```

```python
files = res
```

```python
df =df1
from tqdm import tqdm
for file in tqdm(files):
    # combining multiple excel worksheets
    # into single data frames
    df = pd.concat([df, pd.read_csv(dir_path+"/"+file,index_col="Unnamed: 0")],
                   ignore_index=True, sort=False)
```

```python
df_Eth = df
df_Eth.tail()
```

```python
df_Eth['Timestamp']=pd.to_datetime(df_Eth['Timestamp'])
df_Eth.dtypes
```

```python
df_Eth=df_Eth.sort_values(by='Timestamp')
df_Eth
```

```python
df_Eth.to_csv("Eth.csv")
```
