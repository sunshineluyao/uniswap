```python
import os

# folder path
dir_path = r'/content/drive/MyDrive/all_pool_uniswap/Eth/2022/February'

# list file and directories
res = os.listdir(dir_path)
print(res)
```

```python
df1 = pd.read_excel(dir_path+"/"+"uniswap_ethereum_2022_02_01.xlsx")
df1.head()
```

```python
df1.tail()
```

```python
files = res
```

```python
df =df1
from tqdm import tqdm
for file in tqdm(files): 
    # combining multiple excel worksheets 
    # into single data frames
    df = pd.concat([df, pd.read_excel(dir_path+"/"+file)],
                   ignore_index=True, sort=False)
```

```python
df_Eth_Feb = df
df_Eth_Feb.tail()
```

```python
df_Eth_Feb.to_csv("Eth_Feb.csv")
```
