```python
import os

# folder path
dir_path = r'/content/drive/MyDrive/all_pool_uniswap/Eth/2021/November'

# list file and directories
res = os.listdir(dir_path)
print(res)
```

```python
df1 = pd.read_excel(dir_path+"/"+"uniswap_ethereum_2021_11_1.xlsx")
df1.head()
```

```python
df1.tail()
```

```python
files = res
```

```python
df =df1
from tqdm import tqdm
for file in tqdm(files): 
    # combining multiple excel worksheets 
    # into single data frames
    df = pd.concat([df, pd.read_excel(dir_path+"/"+file)],
                   ignore_index=True, sort=False)
```

```python
df_Eth_Nov2021 = df
df_Eth_Nov2021.tail()
```

```python
df_Eth_Nov2021.to_csv("Eth_Nov2021.csv")
```
